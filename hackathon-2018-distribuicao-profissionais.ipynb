## Quantidade de servidores por unidade

```python
qtd_unidade = df.transpose()

#transforma a primeira linha em indice de colunas
qtd_unidade.columns = qtd_unidade.iloc[0]

#remove a primeira linha
qtd_unidade = qtd_unidade.iloc[1:]

# remove a ultima linha(totais)
qtd_unidade = qtd_unidade.iloc[0:-1]

#olha o fim do dataset
qtd_unidade.tail()
```

```python
qtd_unidade.columns
```

```python
qtd_unidade_geral = qtd_unidade['Total Geral'].sort_values(ascending=False)
top_X = qtd_unidade_geral[0:10]
top_X
```

```python
top_X.plot.bar(x=top_X.index)
```

## Quantidade de servidores por profissão

### Em todas as unidades

```python
### Removendo total
df_sem_total = df.iloc[0:-1]
```

```python
qtd_total = df_sem_total[['PROFISSIONAL','Total Geral']].sort_values(ascending=False,by='Total Geral')
top_X = qtd_total[0:14]
top_X
```

```python
top_X.plot.bar(x=top_X['PROFISSIONAL'])
```

### Na unidade que mais tem servidores

```python
qtd_total = \
df_sem_total[['PROFISSIONAL','SUPERINTENDENCIA DA REGIAO DE SAUDE SUDOESTE                ']]\
.sort_values(ascending=False,by='SUPERINTENDENCIA DA REGIAO DE SAUDE SUDOESTE                ')
top_X = qtd_total[0:14]

top_X
```

```python
top_X.plot.bar(x=top_X['PROFISSIONAL'])
```
